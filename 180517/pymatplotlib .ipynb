## line formatted with title, axis

```python
times = [0, 0.25, 0.5, 0.75]
plt.plot(times, [0, 0.5, 1, 1.2], 'g--')
plt.plot(times, [1, 2, 3, 4], 'r')
plt.title('Concentration of Chlorine vs Time')
plt.ylabel('Concentration (%)')
plt.xlabel('Time (s)')
plt.show()
```

## add a legend

```python
import matplotlib.patches as mpatches
times = [0, 0.25, 0.5, 0.75]
plt.plot(times, [0, 0.5, 1, 1.2], 'g--')
plt.plot(times, [1, 2, 3, 4], 'r')
plt.title('Concentration of Chlorine vs Time')
plt.ylabel('Concentration (%)')
plt.xlabel('Time (s)')

green_patch = mpatches.Patch(color = 'green', label = 'some data')
blue_patch = mpatches.Patch(color = 'blue', label = 'other data')
plt.legend(handles = [green_patch, blue_patch], loc = 'upper left')

plt.show()
```

## 히스토그램


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mu, sigma = 100, 15
x = mu + sigma*np.random.randn(10000) # 기준 100, 시그마15, 10000개 
print(x)
n, bins, patches = plt.hist(x, 50, density=1, facecolor ='g', alpha = 0.7)
plt.axis([40,160,0,0.03])
plt.show()
```

```python
x1 = np.linspace(0.0, 5.0)
print(x1)
x2 = np.linspace(0.0, 2.0)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.subplot(2,1,1)
plt.plot(x1,y1,'yo-')
plt.title('a tale of 2 subplots')
plt.ylabel('damped oscillation')

plt.subplot(2,1,2)
plt.plot(x2,y2, 'r.-')
# plt.title('cos Function')
plt.xlabel('Time (s)')
plt.ylabel('undamped')
plt.show()
```

## Multiple axes

```python
fig, ax1 = plt.subplots()
t = np.arange(0.01,10.0, 0.01)
print(t)
s1 = np.exp(t)
ax1.plot(t,s1,'b-')
ax1.set_xlabel('time (s)') # subplots
ax1.set_ylabel('exp', color='b')
for t1 in ax1.get_yticklabels():
    t1.set_color('b')

ax2 = ax1.twinx() #add subplots
s2 = np.sin(2*np.pi*t)
ax2.plot(t,s2,'r.')
ax2.set_ylabel('sin', color ='r')
for t1 in ax2.get_yticklabels():
    t1.set_color('r')
    
plt.show()

```
